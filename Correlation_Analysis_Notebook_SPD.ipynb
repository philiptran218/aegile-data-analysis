# Correlation Analysis Notebook

This notebook calculates the correlation between selected fields in a CSV file and visualizes the correlation matrix.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Step 1: Load the CSV File

```python
# Replace 'your_file.csv' with your CSV file path
file_path = 'datasets/Student_Performance_Data(SPD24).xlsx'

# Load the dataset
data = pd.read_excel(file_path)

# Display the first few rows of the dataset
data.head()
```

## Step 2: Select Fields for Correlation Analysis

```python
# Specify the fields for correlation analysis
selected_fields = ['Attendance Rate', 'Study Hours', 'Previous Academic Performance', 'Hours of Sleep', 'Homework Completion Rate', 'Reading Proficiency', 
                   'Math Proficiency', 'Science Proficiency', 'Language Proficiency']  # Update with your column names
subset_data = data[selected_fields]

# Display the selected data
subset_data.head()
```

## Step 3: Compute the Correlation Matrix

```python
# Compute the correlation matrix
correlation_matrix = subset_data.corr()

# Display the correlation matrix
correlation_matrix
```

## Step 4: Visualize the Correlation Matrix

```python
# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
